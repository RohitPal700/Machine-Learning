```python
import numpy as np 
import pandas as pd 

#Create a simple data set with missing values
data ={
    'A' :[1,2,100,4,5],
    'B': [3,np.nan,7,8,9],
    'C':[np.nan,12,13,14,15],
    'D':[16,17,18,np.nan,20]
}

#convert the dictionary into pandas dataframe
df=pd.DataFrame(data)
print("original Data Frame")
print(df)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna() # Remove all row those have Nan
```

```python
df.dropna(axis=1) # Remove those column those have Nan
```

```python
df['B'].fillna(df['B'].mean()) # Here I want to fill missing value Nan >> with mean of column 
```

```python
df['B'].mean()
```

```python
df['C'].fillna(13) # here missing value is replaced by random number 
```

```python
# whenever you have normal distribution data , impute with mean 
```

```python
# here we fill missing valuse of age  by mean 

df['mean_imputation']=df['age'].fillna(df['age'].mean())
```

```python
df[['age','mean_imputation']]
```

```python
sns.displot(df['mean_imputation'],kde=True)
```

```python
# mean imputation  work well with normal distribution data / no outlier is present in data 
```
